```python
# Iterator 範例

numbers = [10,20,30]

it = iter(numbers)

print(next(it))
print(next(it))
print(next(it))
```

```python
# Generator 範例

def my_generator():

    for i in range(5):

        yield i


for v in my_generator():

    print(v)
```

```python
# itertools.islice

import itertools

data = range(100)

for v in itertools.islice(data,10):

    print(v)
```

```python
# infinite generator

def infinite():

    i = 0

    while True:

        yield i

        i += 1


import itertools

for v in itertools.islice(infinite(),10):

    print(v)
```

```python
# chain

import itertools

a = [1,2]
b = [3,4]

for v in itertools.chain(a,b):

    print(v)
```
